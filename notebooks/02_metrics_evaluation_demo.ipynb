# Segmentation Metrics Evaluation Demo

## Overview
This notebook demonstrates how segmentation quality can be evaluated using standard quantitative metrics:
- Dice Similarity Coefficient (DSC)
- Intersection over Union (IoU)
- Hausdorff Distance (conceptual placeholder)

These metrics were used in the original project to compare the baseline and improved workflows for intracranial haemorrhage segmentation.

```python
import numpy as np
import pandas as pd
```

## Example segmentation masks

Below is a simple binary mask example showing a ground-truth region and a predicted region.
This is only for metric demonstration.

```python
y_true = np.zeros((10, 10), dtype=int)
y_pred = np.zeros((10, 10), dtype=int)

y_true[2:6, 3:7] = 1
y_pred[3:7, 4:8] = 1

print("Ground truth mask:")
print(y_true)
print("\nPredicted mask:")
print(y_pred)
```

## Dice and IoU functions

```python
def dice_score(y_true, y_pred):
    intersection = np.sum(y_true * y_pred)
    return (2 * intersection) / (np.sum(y_true) + np.sum(y_pred) + 1e-8)

def iou_score(y_true, y_pred):
    intersection = np.sum(y_true * y_pred)
    union = np.sum((y_true + y_pred) > 0)
    return intersection / (union + 1e-8)

dice = dice_score(y_true, y_pred)
iou = iou_score(y_true, y_pred)

print(f"Dice score: {dice:.4f}")
print(f"IoU score: {iou:.4f}")
```
