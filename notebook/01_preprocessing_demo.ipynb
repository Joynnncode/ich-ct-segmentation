# Preprocessing Demo for Intracranial Haemorrhage Segmentation

## Overview
This notebook presents a simplified demonstration of the preprocessing logic used in an intracranial haemorrhage (ICH) segmentation workflow for non-contrast CT images.

The original project explored whether preprocessing refinement could improve segmentation quality, particularly for small haemorrhage lesions.

## Main preprocessing steps
- Intensity normalisation
- Resampling
- Region-of-interest style cropping
- Visual comparison before and after preprocessing


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create a simple placeholder CT slice

The original clinical dataset is not included in this repository due to data restrictions.  
This notebook uses a synthetic image to demonstrate preprocessing logic only.

```python
np.random.seed(42)

image = np.random.normal(loc=40, scale=20, size=(256, 256))
image[110:145, 120:155] += 55  # simulate brighter lesion-like region

plt.figure(figsize=(5, 5))
plt.imshow(image, cmap="gray")
plt.title("Synthetic CT-like Slice (Before Preprocessing)")
plt.axis("off")
plt.show()
```

## Intensity normalisation

In CT workflows, normalisation can help standardise image intensity ranges before downstream analysis.
This is especially useful when image appearance varies across scans.

```python
def normalize_image(img):
    img_min = img.min()
    img_max = img.max()
    return (img - img_min) / (img_max - img_min + 1e-8)

normalized_image = normalize_image(image)

plt.figure(figsize=(5, 5))
plt.imshow(normalized_image, cmap="gray")
plt.title("After Intensity Normalisation")
plt.axis("off")
plt.show()
```

## Cropping a region of interest

Small haemorrhage lesions can be difficult to segment accurately.
Cropping or patch-based strategies can help the model focus more closely on informative regions.

```python
cropped_image = normalized_image[90:170, 100:180]

plt.figure(figsize=(5, 5))
plt.imshow(cropped_image, cmap="gray")
plt.title("Cropped Region of Interest")
plt.axis("off")
plt.show()
```

## Before and after comparison

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(image, cmap="gray")
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(normalized_image, cmap="gray")
axes[1].set_title("Normalised")
axes[1].axis("off")

axes[2].imshow(cropped_image, cmap="gray")
axes[2].set_title("Cropped ROI")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
