# Results Visualisation

## Overview
This notebook visualises the main quantitative results from the intracranial haemorrhage segmentation project.

The goal is to present the workflow outcome clearly and concisely in a portfolio-friendly format.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.DataFrame({
    "Model": ["Baseline 3D U-Net", "3D Improved U-Net", "2D Improved U-Net"],
    "DSC": [0.6047, 0.6667, 0.6804],
    "HD": [49.9967, 41.6946, 29.3923],
    "IoU": [0.5052, 0.5686, 0.5865]
})

results
```

## Dice score comparison

```python
plt.figure(figsize=(7, 4))
plt.bar(results["Model"], results["DSC"])
plt.ylabel("DSC")
plt.title("Dice Score Comparison")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 4))
plt.bar(results["Model"], results["HD"])
plt.ylabel("HD")
plt.title("Hausdorff Distance Comparison")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

## Key takeaway

The improved workflows achieved better overlap-based metrics than the baseline model, with lower Hausdorff Distance and stronger lesion delineation performance.

This supports the conclusion that workflow refinement and preprocessing can improve segmentation quality, particularly for challenging small haemorrhage lesions.
